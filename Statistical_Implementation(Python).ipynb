## Variance-Function 

```python
def var(d):
    
    var=0
    l1=[]
    l=len(d)
    Sum=0
    for i in d:
        Sum=Sum+i
    mean=Sum/l
    
    for i in d:
        l1.append((mean-i)**2)
    
    s=sum(l1)
    var=s/(l-1)
    std=var**0.5
    
    print(f"Variance:{var} and Standard-Deviation:{std}")
    
```

```python
var(data_copy)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(data)
```

```python
plot_data(s)
```

```python
plot_data(p)
```

```python
plot_data(np.log(p))
```
